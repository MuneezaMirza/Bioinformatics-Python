To download FASTQ file from FASTA file.


```python
import Bio
from Bio import SeqIO
```

```python
input_fasta = "gene.fna"
output_fastq = "gene.fna.fastq"
```

```python
with open(output_fastq, "w") as out_file:
    for record in SeqIO.parse(input_fasta, "fasta"):
        record.letter_annotations["phred_quality"] = [40] * len(record.seq)
        SeqIO.write(record, out_file, "fastq")

```
